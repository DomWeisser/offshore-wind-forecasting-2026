```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LongitudeFormatter, LatitudeFormatter
from matplotlib.patheffects import withStroke
```

```python
sites = [
    {'name': 'Beatrice', 'lat': 58.254, 'lon': -2.885},
    {'name': 'Hornsea', 'lat': 53.885, 'lon': 1.791},
    {'name': 'Walney', 'lat': 54.088, 'lon': -3.738},
    {'name': 'Gode Wind', 'lat': 54.04, 'lon': 7.00, 'dx': -0.8, 'dy': 0.2},
    {'name': 'Baltic Eagle', 'lat': 54.83, 'lon': 13.80, 'dx': -2.0, 'dy': 0.2},  
    {'name': "les d'Yeu\net de Noirmoutier", 'lat': 46.87, 'lon': -2.51},
]

fig = plt.figure(figsize=(12, 10))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_extent([-5, 15, 45, 60], crs=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND, facecolor='lightyellow')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')

ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle='--')

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=0.5,
                  color='gray', alpha=0.5, linestyle='--')
gl.top_labels = False
gl.right_labels = False
gl.xformatter = LongitudeFormatter()
gl.yformatter = LatitudeFormatter()
gl.xlabel_style = {'size': 10}
gl.ylabel_style = {'size': 10}

for site in sites:

    ax.plot(site['lon'], site['lat'], marker='o', color='red', markersize=8,
            transform=ccrs.PlateCarree())


    dx = site.get('dx', 0.2)
    dy = site.get('dy', 0.2)


    ax.text(site['lon'] + dx, site['lat'] + dy, site['name'],
            fontsize=22, fontweight='bold', transform=ccrs.PlateCarree(),
            path_effects=[withStroke(linewidth=3, foreground='white')])


plt.savefig(r'...', dpi=300, bbox_inches='tight')

plt.show()
```
